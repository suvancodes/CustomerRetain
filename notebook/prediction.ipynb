```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
import pickle
```

```python
## load the model ,scaler and one hot encoder pikkel files
model = load_model('artifect/model.h5')

with open('artifect/scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
    
with open('artifect/level_encoder_gender.pkl', 'rb') as file:
    level_encoder_gender = pickle.load(file)
    
with open('artifect/onehot_encoder_geo.pkl', 'rb') as file:
    onehot_encoder_geo = pickle.load(file)
```

```python
# CreditScore	Gender	Age	Tenure	Balance	NumOfProducts	HasCrCard	IsActiveMember	EstimatedSalary	Exited	Geography_France	Geography_Germany	Geography_Spain
input_data = {
    'CreditScore': 25,
    'Gender': 'Female',
    'Age': 45,
    'Tenure': 5,
    'Balance': 10000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000,
    'Geography': 'France'
}
```

```python
input_data['Gender'] = level_encoder_gender.transform([input_data['Gender']])[0]
```

```python
geo_encoded = onehot_encoder_geo.transform([[input_data['Geography']]])
geo_encoded_df = pd.DataFrame(geo_encoded, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_data = pd.DataFrame([input_data])
```

```python
input_data = input_data.drop('Geography', axis=1)
input_data = pd.concat([input_data, geo_encoded_df], axis=1)
```

```python
input_data = scaler.transform(input_data)
```

```python
# get prediction
prediction = model.predict(input_data)
```

```python
prediction_proba = prediction[0][0]
prediction_proba
```

```python
if prediction_proba>0.5:
    print("the custormer is likely to churn")
else:
    print("the custormer is not likely to churn")
```
